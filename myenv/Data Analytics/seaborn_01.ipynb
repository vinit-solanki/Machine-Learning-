```python
import seaborn as sns
print(sns.__version__)
```

```python
tips = sns.load_dataset('tips')
```

```python
# Scatterplot
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.title("tip vs total_bill")
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lineplot(x='size', y='tip', data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
# Categorical Plots
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar Plot of Total bill by day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box Plot of Total bill by day")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot of Total bill by day")
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'], fill=True, color='r')
plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
# PairPlot
sns.pairplot(tips)
plt.title('Pair Plot of Tips Data')
plt.show()
```

```python
tips_numeric = tips.select_dtypes(include='number')
print(tips_numeric.corr())
sns.heatmap(tips_numeric.corr(), annot=True)
```
