```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [1,5,6,8,3]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
# plt.show()
```

```python
# creating a customized lined plot
plt.plot(x,y,color='red',linestyle=':',marker='o',linewidth=2,markersize=4)
```

```python
# Multiple plot
x = [1,2,3,4,5]
y1 = [2,3,6,8,9]
y2 = [3,5,7,7,8]

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('X vs Y1')
plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title('X vs Y2')
#(row,col,sub-row)
plt.subplot(2,2,3) 
plt.plot(x,y1,color='green')
plt.plot(x,y2, color='red')
plt.title('X vs Y1,Y2')
```

```python
# Bar plot
categories = ['A','B','C','D','E']
values = [5,7,8,6,9]
plt.bar(categories,values,color='purple')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Histogram
data = [1,2,3,4,5,5,2,8,9,14,7,32,67]
plt.hist(data, bins=4,edgecolor='black')
plt.hist(data,bins=5,edgecolor='black')
```

```python
# Scatterplot
plt.scatter(x,y,color='blue',marker='x')
```

```python
# pie chart
labels = ['A','B','C','D']
sizes = [30,50,50,40]
colors = ['gold','yellow','red','purple']
explode = (0,0.1,0.1,0)
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.2f%%",explode=explode,shadow=True)
```
