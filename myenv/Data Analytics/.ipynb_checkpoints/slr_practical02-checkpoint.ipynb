```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HeightWeight.csv')
```

```python
df = df.rename(columns={ "Height(Inches)": "Height", "Weight(Pounds)": "Weight" })
df.columns
```

```python
df.head()
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
print(df.isnull().sum())
print(df.isna().sum())
```

```python
# Independent feature
X = df['Weight']
X.shape
```

```python
y = df['Height']
y.shape
```

```python
X = pd.DataFrame(X)
X = X.apply(pd.to_numeric, errors='coerce')
```

```python
y = pd.DataFrame(y)
y = y.apply(pd.to_numeric, errors='coerce')
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_series = df['Weight']
print(X_series)
print(np.array(X_series).shape)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train, y_train)
```

```python
# Coef_ -> Slope of the best fit line
regression.coef_
```

```python
# Intercept of the best fit line
regression.intercept_
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, regression.predict(X_test), color="red", linewidth=1)
```

```python
# Prediction of test data
y_pred = regression.predict(X_test)
```

```python
# Performance Metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mae)
print(mse)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
score
```
